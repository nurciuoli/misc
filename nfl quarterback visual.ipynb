```python
import pandas as pd

df = pd.read_csv('nflpassingstats_10_7_24.csv')
df = df.iloc[:-5]
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from adjustText import adjust_text
sns.set_style("darkgrid")
fig,ax = plt.subplots(figsize =(15,15))

sns.scatterplot(data = df,y='Yds/Att',x='Rate')
texts = []
for line in range(0,df.shape[0]):
     texts.append(ax.text(df.Rate[line], df['Yds/Att'][line], df.Player[line]))

adjust_text(texts)

plt.title('NFL QB Passer Rating vs. Yards per Attempt through 10/7/23')
plt.xlabel('Passer Rating')
plt.savefig('qb_passer_ratings.png')
plt.show()
```
